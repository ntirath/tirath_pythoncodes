```python
set1=set()
set2=set()
for i in range(5):
    set1.add(i)

for i in range(3,9):
    set2.add(i)

set3 = set1.intersection(set2)
print("Intersection of sets")
print(set3)

set3 = set1 & set2
print("Intersection of sets using & operator")
print(set3)
```

```python
#difference
set1 = set()
set2 = set()
for i in range(5):
    set1.add(i)

for i in range(3,9):
    set2.add(i)

set3 = set1.difference(set2)
print(set3)

set3 = set1 - set2
print(set3)

set3 = set2 - set1
print(set3)
```
